## Indeksy i przycinanie
* Indexing and slicing (indeksowanie i krajanie na plastry(Google translate)/ziomki na plastry/plasterkowanie)

Indeksowanie i przynacie macierzy odbywa się w ten sam sposób co w przypadku prostych typów sekwencyjnych

```python
a = np.arange(10)
print(a)
print(a[0])
print(a[::-1]) # odwrotnie
print(a[2:6:2]) # od 2 do 6 ze skokiem 2
```

```python
np.set_printoptions(precision=2)

a=np.random.rand(3,3)
print(a)
print(a[(1,1)]) # trzy sposoby
print(a[1]) # sam wiersz
print(a[:,0]) # sama kolumna
```

<b> Wczytywanie obrazów </b>

```python
from PIL import Image
```

```python
obraz = Image.open("Lena.png")
```

```python
obraz
```

```python
np_obraz = np.array(obraz)
```

```python
np_obraz.shape
```

```python
nowy_obraz = np_obraz[200:300]
Image.fromarray(nowy_obraz)
```

```python
nowy_obraz = np_obraz[240:290,250:350]
Image.fromarray(nowy_obraz)
```

```python
nowy_obraz = np_obraz[:,:,2]
Image.fromarray(nowy_obraz)
```

```python
print(obraz.format)
print(obraz.size)
print(obraz.mode)
```

```python
#osobne okno
obraz.show()
```

```python
#Zapis
obraz.save("Lena2.jpg","JPEG")
```

Rozbijanie kanałów RGB

```python
nowy_obraz = np.array(obraz)
R = nowy_obraz[:,:,0]
G = nowy_obraz[:,:,1]
B = nowy_obraz[:,:,2]
Image.fromarray(B)
```

Złożenie kanałów

```python
nowy_obraz = np.stack((B,G,G),axis=2)
Image.fromarray(nowy_obraz)
```

Generowanie losowych obrazów

```python
obraz = np.eye(512)
obraz = Image.fromarray(obraz*255).convert('L')
obraz
```

```python
obraz = np.eye(512)
obraz[:,100:150] = 1
obraz[100:200,200:250] = 1
obraz[200:210,:] = 1
Image.fromarray(obraz*255).convert('L')
```
